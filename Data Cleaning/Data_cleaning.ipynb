```python
data=pd.read_csv('Project/tennis_atp-master/atp_players.csv')
wta_data=pd.read_csv('Project/tennis_wta-master/wta_players.csv')
wta_data['player_id']= wta_data['player_id']+1000000
data=pd.concat([data,wta_data], axis=0)
data['name'] = data['name_first'].str.cat(data['name_last'], sep=' ')

data
```

```python
# Function to convert numeric date format to string, then to datetime, handling NaN values
def convert_to_date(num):
    try:
        if pd.isna(num):
            return np.nan
        else:
            num_str = str(int(float(num)))  # Ensure the number is properly converted to string
            return pd.to_datetime(num_str, format='%Y%m%d')
    except (ValueError, TypeError):
        return np.nan


# Apply the function to the date_column
data['dob'] = data['dob'].apply(convert_to_date)

# Format the date column as desired
data['dob'] = data['dob'].dt.strftime('%Y-%m-%d')

data['is_atp'] = data['player_id'] < 1000000
data.loc[data['name'] == 'E Cliff', 'player_id'] += 1000000


data[["player_id", "name", "hand", "dob", "ioc", "height", "is_atp"]].to_csv('CIS-5500/player_data.csv', index=False) 
data
```

```python
def convert_to_date(num):
    try:
        if pd.isna(num):
            return np.nan
        else:
            num_str = str(int(float(num)))  # Ensure the number is properly converted to string
            return pd.to_datetime(num_str, format='%Y%m%d')
    except (ValueError, TypeError):
        return np.nan
```

```python
#processed data

# List all files in the current directory
atp_directory = 'Project/tennis_atp-master'
wta_directory = 'Project/tennis_wta-master'

atp_files = os.listdir(atp_directory)
wta_files = os.listdir(wta_directory)

# Filter files that match the pattern "atp_matches_{yyyy}.csv"
atp_files = [file for file in atp_files if file.startswith('atp_matches_') and file.endswith('.csv')]
wta_files = [file for file in wta_files if file.startswith('wta_matches_') and file.endswith('.csv')]

# Read and concatenate all matching files into a single dataframe
atp_dataframes = [pd.read_csv(os.path.join(atp_directory, file)) for file in atp_files]
wta_dataframes = [pd.read_csv(os.path.join(wta_directory, file)) for file in wta_files]

atp = pd.concat(atp_dataframes, ignore_index=True)
wta = pd.concat(wta_dataframes, ignore_index=True)
wta['winner_id']= wta['winner_id']+1000000

wta['loser_id']= wta['loser_id']+1000000

combined_df = pd.concat([atp, wta], ignore_index=True)

# Apply the function to the date_column
combined_df['tourney_date'] = combined_df['tourney_date'].apply(convert_to_date)

# Format the date column as desired
combined_df['tourney_date'] = combined_df['tourney_date'].dt.strftime('%Y-%m-%d')

combined_df['tourney_year'] = combined_df['tourney_id'].str[:4]
combined_df['tourney_id'] = combined_df['tourney_id'].str[5:]
# Display the first few rows of the combined dataframe
combined_df.head()

```

```python
matches= combined_df[combined_df['tourney_date'] > "2014-01-01"]
winner_entry_stats = matches[['tourney_id', 'tourney_date', 'winner_id', 'winner_rank', 'winner_rank_points', 'winner_seed','tourney_year']].rename(columns={
    'winner_id': 'player_id',
    'winner_rank': 'entry_rank',
    'winner_rank_points': 'entry_rank_points',
    'winner_seed':'seed'
})

loser_entry_stats = matches[['tourney_id', 'tourney_date', 'loser_id', 'loser_rank', 'loser_rank_points','loser_seed']].rename(columns={
    'loser_id': 'player_id',
    'loser_rank': 'entry_rank',
    'loser_rank_points': 'entry_rank_points',
    'loser_seed':'seed'
})

# Combine entry stats
all_entry_stats = pd.concat([winner_entry_stats, loser_entry_stats])

# Get the first occurrence of entry statistics for each player in each tournament
entry_stats = all_entry_stats.drop_duplicates(subset=['player_id', 'tourney_id', 'tourney_date'])

# Merge entry statistics with player statistics
#player_stat_at_tourney = pd.merge(player_stats, entry_stats, on=['player_id', 'tourney_id', 'tourney_year'])

```

```python
player_stat_with_dob  = pd.merge(entry_stats, data[['player_id', 'dob']], on='player_id', how='left')
def calculate_age(dob, tourney_year):
    return tourney_year - dob.year

# Calculate age at tournament entry
player_stat_with_dob['dob'] = pd.to_datetime(player_stat_with_dob['dob'], format='%Y-%m-%d')
player_stat_with_dob['tourney_date'] = pd.to_datetime(player_stat_with_dob['tourney_date'], format='%Y-%m-%d')
player_stat_with_dob['tourney_year_from_date'] = player_stat_with_dob['tourney_date'].dt.year
player_stat_with_dob['tourney_year'] = player_stat_with_dob['tourney_year'].fillna(player_stat_with_dob['tourney_year_from_date']).astype(int)

player_stat_with_dob['tourney_year'] = player_stat_with_dob['tourney_year'].astype(int)

player_stat_with_dob['age'] = player_stat_with_dob.apply(lambda row: calculate_age(row['dob'], row['tourney_year']), axis=1)
```

```python
player_stat_with_dob.fillna(0, inplace = True)
player_stat_with_dob['age'] = player_stat_with_dob['age'].astype(int)
player_stat_with_dob['entry'] = player_stat_with_dob['entry_rank'].astype(int)
player_stat_with_dob['seed'] = pd.to_numeric(player_stat_with_dob['seed'], errors='coerce')

player_stat_with_dob['seed'] = player_stat_with_dob['seed'].astype(int)
```
