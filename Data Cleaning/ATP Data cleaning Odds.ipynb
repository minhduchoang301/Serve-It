```python
import pandas as pd
import os

import fuzzywuzzy
from fuzzywuzzy import process
```

# Load all Odds data for ATP

```python

# Define the directory containing the Excel files
directory = 'CIS 5500/Project/Odds data/ATP/'

# Create an empty list to store dataframes
dataframes = []

# Loop through each file in the directory
for filename in os.listdir(directory):
    if filename.endswith('.xlsx'):
        # Load the Excel file
        df = pd.read_excel(os.path.join(directory, filename))
        # Append the dataframe to the list
        dataframes.append(df)

# Concatenate all dataframes into one
all_matches_df = pd.concat(dataframes, ignore_index=True)

# Display the first few rows of the concatenated dataframe
all_matches_df.head()
```

# First Attempt to Match Player Data to get to Player ID

The one whose last name with # entry > 1 are temporarily excluded

```python
player_data = pd.read_csv('CIS 5500/CIS-5500/player_data.csv')
atp_player_data = player_data[player_data['is_atp']==True]
atp_player_data
```

```python
# Function to extract last name and first initial
def extract_name(name):
    parts = name.split()
    last_name = parts[0]
    first_initial = parts[1][0]
    return last_name, first_initial

# Function to find the best match
def find_best_match(name, choices):
    match, score = process.extractOne(name, choices)
    return match if score > 80 else None  # Adjust the threshold as needed

# Add columns for winner and loser IDs
all_matches_df['winner_id'] = None
all_matches_df['loser_id'] = None
atp_player_data = atp_player_data.dropna(subset=['name'])

# Create a dictionary for quick look-up of player IDs
player_name_to_id = {row['name']: row['player_id'] for _, row in atp_player_data.iterrows()}

# Create a dictionary to map last names to full names
last_name_to_names = {}
for name in player_name_to_id.keys():
    last_name = name.split()[-1]
    if last_name not in last_name_to_names:
        last_name_to_names[last_name] = []
    last_name_to_names[last_name].append(name)

# Loop through each match
for index, row in all_matches_df.iterrows():
    # Extract winner's last name and first initial
    winner_last, winner_initial = extract_name(row['Winner'])
    # Construct the full name pattern for matching
    winner_pattern = winner_last + ' ' + winner_initial
    
    # Extract loser's last name and first initial
    loser_last, loser_initial = extract_name(row['Loser'])
    # Construct the full name pattern for matching
    loser_pattern = loser_last + ' ' + loser_initial
    
    # Match winner based on last name
    if winner_last in last_name_to_names and len(last_name_to_names[winner_last]) == 1:
        best_winner_match = last_name_to_names[winner_last][0]
        all_matches_df.at[index, 'winner_id'] = player_name_to_id[best_winner_match]
    
    # Match loser based on last name
    if loser_last in last_name_to_names and len(last_name_to_names[loser_last]) == 1:
        best_loser_match = last_name_to_names[loser_last][0]
        all_matches_df.at[index, 'loser_id'] = player_name_to_id[best_loser_match]

# Display the updated dataframe
all_matches_df
```

```python
def get_tourney_level(series):
    if series == 'Grand Slam':
        return 'G'
    elif series in ['Masters 1000', 'Masters Cup']:
        return 'M'
    elif series in ['ATP250', 'ATP500']:
        return 'A'
    else:
        return None

# Apply the function to create the 'tourney_level' column
all_matches_df['tourney_level'] = all_matches_df['Series'].apply(get_tourney_level)
```

```python
# Display unique values in 'Series' and 'Court' columns
unique_series = all_matches_df['Series'].unique() if 'Series' in all_matches_df.columns else []
unique_court = all_matches_df['Court'].unique() if 'Court' in all_matches_df.columns else []
unique_surface = all_matches_df['Surface'].unique() if 'Surface' in all_matches_df.columns else []

print("Unique values in 'Series':", unique_series)
print("Unique values in 'Court':", unique_court)
print("Unique values in 'Surface':", unique_surface)
```

```python
all_matches_df['Date'] = pd.to_datetime(all_matches_df['Date'], format='%Y-%m-%d')
```

```python
all_matches_df['Year'] = all_matches_df['Date'].dt.year
```

```python
all_matches_df.rename(columns={
    'Surface': 'surface',
    'WRank': 'winner_rank',
    'LRank': 'loser_rank'
}, inplace=True)
```

# Second Attempt: Fuzzy Match to Tourney Match and backfill the missing ID

Match to tourney match based on tourney name, tourney level, surface, tourney_year, winner_rank and loser rank

By this, we can backfill the winner or loser id missing as well

```python
tourney_match = pd.read_csv('CIS 5500/CIS-5500/tourney_match.csv')
```

```python
unique_series = tourney_match['tourney_level'].unique() if 'tourney_level' in tourney_match.columns else []
unique_surface = tourney_match['surface'].unique() if 'surface' in tourney_match.columns else []

print("Unique values in 'Series':", unique_series)
print("Unique values in 'Surface':", unique_surface)
```

```python
tourney_match
```

```python
# match on tourney_level, surface, contains (tourney_name), winner_rank, loser_rank -> tourney_id, tourney_name, surface, tourney_level, tourney_year, match_num, winner_id, loser_id, winner_rank, loser_rank
```

```python
# Perform the merge with the specified conditions
merged_df = pd.merge(
    all_matches_df,
    tourney_match,
    left_on=['tourney_level', 'surface', 'Year', 'winner_rank', 'loser_rank'],
    right_on=['tourney_level', 'surface', 'tourney_year', 'winner_rank', 'loser_rank'],
    suffixes=('_all_matches', '_tourney')
)

# Filter rows where 'Tournament' contains 'tourney_name'
merged_df = merged_df[merged_df.apply(lambda row: row['tourney_name'].lower() in row['Tournament'].lower(), axis=1)]
```

```python
# Fill winner_id_all_matches with winner_id_tourney if it is None
merged_df['winner_id_all_matches'].fillna(merged_df['winner_id_tourney'], inplace=True)

# Fill loser_id_all_matches with loser_id_tourney if it is None
merged_df['loser_id_all_matches'].fillna(merged_df['loser_id_tourney'], inplace=True)
```
